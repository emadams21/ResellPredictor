```python
#importing the libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
sneakerData = pd.read_csv('data/parsed_data.csv', index_col = 0)
```

```python
sneakerData['Black'] = sneakerData['Black'].fillna(0)
sneakerData['Red'] = sneakerData['Red'].fillna(0)
sneakerData['White'] = sneakerData['White'].fillna(0)
sneakerData['Green'] = sneakerData['Green'].fillna(0)
sneakerData['Neo'] = sneakerData['Neo'].fillna(0)
sneakerData['Grey'] = sneakerData['Grey'].fillna(0)
sneakerData['Orange'] = sneakerData['Orange'].fillna(0)
sneakerData['Tan/Brown'] = sneakerData['Tan/Brown'].fillna(0)
sneakerData['Pink'] = sneakerData['Pink'].fillna(0)
sneakerData['Blue'] = sneakerData['Blue'].fillna(0)
sneakerData['Colorful'] = sneakerData['Colorful'].fillna(0)
```

```python
sneakerData
```

```python
sneakerData.info()
```

```python
 GroupedShoes = sneakerData.groupby(['Sneaker Name','Shoe Size'],as_index=False).agg({'Retail Price': 'min' , 'Sale Price': 'mean','Release Date': 'first','Order Date': 'first','Red':'first','Black':'first','White':'first','Green':'first','Neo':'first','Orange':'first','Tan/Brown':'first','Pink':'first','Blue':'first','Colorful':'first'})
```

```python
GroupedShoes['Shoe Size'] = GroupedShoes['Shoe Size'].astype(str)
```

```python
GroupedShoes['Order Date'] = pd.to_datetime(GroupedShoes['Order Date'],infer_datetime_format=True) 
GroupedShoes['Release Date'] = pd.to_datetime(GroupedShoes['Release Date'],infer_datetime_format=True) 
```

```python
GroupedShoes['Order Date'][0].month
```

```python
GroupedShoes
```

```python
GroupedShoes['release_season'] = GroupedShoes['Release Date'].apply(lambda x: (x.month%12 + 3)//3)
```

```python
season_name = {1:'Winter', 2:'Spring', 3:"Summer",4:"Fall"}
GroupedShoes['release_season'] = GroupedShoes['release_season'].map(season_name)
```

```python
GroupedShoes['release_year'] = GroupedShoes['Release Date'].apply(lambda x: x.year)
GroupedShoes['release_season'] = GroupedShoes['release_season'] + ' ' + GroupedShoes['release_year'].astype(str)
```

```python
GroupedShoes
```

```python
GroupedShoes['order_season'] = GroupedShoes['Order Date'].apply(lambda x: (x.month%12 + 3)//3)
```

```python
season_name = {1:'Winter', 2:'Spring', 3:"Summer",4:"Fall"}
GroupedShoes['order_season'] = GroupedShoes['order_season'].map(season_name)
```

```python
GroupedShoes['order_year'] = GroupedShoes['Order Date'].apply(lambda x: x.year)
GroupedShoes['order_season'] = GroupedShoes['order_season'] + ' ' + GroupedShoes['release_year'].astype(str)
```

```python
GroupedShoes = GroupedShoes.rename(columns={"Sale Price":"Resell Price"})
```
